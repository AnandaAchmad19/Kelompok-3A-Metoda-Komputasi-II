## 2. Analisis Fourier

Analisis Fourier merupakan salah satu alat dasar untuk menganalisis dan menyelesaikan persamaan diferensial parsial (PDE) linear. Untuk domain terbatas (finite), kita biasanya menggunakan deret Fourier untuk merepresentasikan solusinya. Penggunaan deret Fourier ini sama dengan memandang fungsi-fungsi eigen dari operator diferensial sebagai basis ruang solusi. Interpretasi aljabar linear terhadap PDE ini sangat kuat dan akan kita jumpai berulang kali.

### Deret Fourier

Deret Fourier dari suatu fungsi memiliki beberapa bentuk tergantung pada jenis fungsi yang dipertimbangkan. Ingat bahwa sin x adalah fungsi ganjil (odd function) dan cos x adalah fungsi genap (even function). Pada interval terbatas, maka:

- Untuk fungsi ganjil, kita dapat menggunakan deret Fourier sinus.
- Untuk fungsi genap, kita dapat menggunakan deret Fourier kosinus.

Jika fungsi tidak termasuk ganjil maupun genap, maka kita menggunakan deret Fourier lengkap, yang sering ditulis sebagai:

$$
f(x) = a_0 + \sum_{n=1}^\infty a_n \cos \left( \frac{n \pi x}{L} \right) + \sum_{n=1}^\infty b_n \sin \left( \frac{n \pi x}{L} \right),
$$

dengan koefisien-koefisien:

$$
\begin{aligned}
a_0 &= \frac{1}{L} \int_0^L f(x) \, dx, \\
a_n &= \frac{2}{L} \int_0^L f(x) \cos \frac{n \pi x}{L} \, dx, \\
b_n &= \frac{2}{L} \int_0^L f(x) \sin \frac{n \pi x}{L} \, dx.
\end{aligned}
$$

Koefisien-koefisien ini merupakan proyeksi dari fungsi f(x) ke dalam basis Fourier yang terdiri dari fungsi cos dan sin, yang membentuk basis ortogonal.

### Transformasi Fourier

Transformasi Fourier dari suatu fungsi $v(x)$ didefinisikan oleh

$$
\widehat{v}(\xi) = \frac{1}{\sqrt{2 \pi}} \int_{-\infty}^{\infty} v(x) e^{-i \xi x} \, dx.
$$

Agar transformasi ini ada (eksist), kita memerlukan fungsi $v$ termasuk dalam ruang $L^2$, yaitu

$$
\|v\|_2 = \left( \int_{-\infty}^{\infty} |v(x)|^2 \, dx \right)^{1/2} < \infty.
$$

Fungsi $\widehat{v}(\xi)$ juga berada dalam $L^2$ karena menurut identitas Parseval berlaku

$$
\|v(x)\|_2 = \|\widehat{v}(\xi)\|_2.
$$

Kita juga dapat menggunakan transformasi invers untuk mendapatkan kembali fungsi asal:

$$
v(x) = \frac{1}{\sqrt{2 \pi}} \int_{-\infty}^{\infty} \widehat{v}(\xi) e^{i \xi x} \, d\xi.
$$


Salah satu sifat paling penting dari Transformasi Fourier yang telah kita gunakan di atas adalah bahwa transformasi Fourier mengubah turunan spasial menjadi hubungan aljabar:

$$
\widehat{v_x} = i \xi \widehat{v}.
$$

Hasil ini muncul karena fungsi-fungsi

$$
e^{i\xi x}
$$

merupakan **fungsi eigen** (eigenfunctions) dari operator

$$
\frac{\partial}{\partial x}.
$$

Sifat ini juga berlaku dalam kasus diskret, di mana kita akan menemukan bahwa analog diskret dari fungsi-fungsi eigen tersebut akan menjadi **vektor eigen** (eigenvectors) dari operator beda hingga (finite difference) yang setara.

Latihan: Cobalah ambil transformasi Fourier dari persamaan adveksi:
$$u_t + c u_x = 0.$$

Karena turunan terhadap waktu tidak terpengaruh oleh transformasi Fourier, kita memperoleh persamaan
$$\widehat{u}_t = -i \xi c \, \widehat{u},$$
yang merupakan sistem tak hingga dari persamaan diferensial biasa (ODE). Sekarang cobalah selesaikan sistem ini untuk menemukan solusi dari persamaan adveksi.

Kita tahu bahwa solusi dari ODE-ODE tersebut berbentuk
$$\widehat{u}(\xi, t) = \widehat{u}(\xi, 0) \, e^{-i \xi c t}.$$
Dengan menggunakan transformasi invers, kita memperoleh
$$\begin{aligned}
u(x,t) &= \frac{1}{\sqrt{2 \pi}} \int_{-\infty}^{\infty} \widehat{u}(\xi, 0) \, e^{-i \xi c t} \, e^{i \xi x} \, d\xi, \\
&= \frac{1}{\sqrt{2 \pi}} \int_{-\infty}^{\infty} \widehat{u}(\xi, 0) \, e^{i \xi (x - c t)} \, d\xi, \\
&= u(x - c t, 0).
\end{aligned}$$
