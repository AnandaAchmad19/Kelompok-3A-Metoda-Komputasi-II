# Tinjauan: Persamaan Differensial Parsial

Tujuan kuliah ini adalah meninjau latar belakang paling dasar untuk diskusi kita tentang metode numerik untuk persamaan differensial parsial (PDE). Sebagian besar yang akan kita gunakan adalah pengetahuan umum tentang klasifikasi PDE, teknik penyelesaian untuk setiap jenis PDE, dan beberapa pengetahuan tentang properti solusi. Ini juga mencakup pertanyaan terkait kebebasan masalah (well-posedness), kondisi batas, dan masalah lainnya.

Pertama beberapa notasi untuk memulai, kita akan menggunakan notasi
$$
    \frac{\partial u}{\partial x} = u_x
$$
untuk menyatakan turunan parsial dari fungsi $u$ terhadap variabel $x$.

Turunan orde lebih tinggi atau turunan campuran dinyatakan serupa sebagai
$$
    \frac{\partial^2 u}{\partial x^2} = u_{xx} \quad \quad \frac{\partial^2 u}{\partial y \partial x} = u_{xy}
$$
menyadari bahwa secara umum urutan turunan dapat berpengaruh.

Kapan kita dapat mengubah urutan turunan dalam kasus-kasus ini?

Di sini kita telah menyoroti variabel yang biasanya terkait dengan ruang Kartesius tetapi juga akan berguna untuk mempertimbangkan turunan koordinat polar seperti
$$
    \frac{\partial u}{\partial \theta} = u_\theta
$$
atau dalam waktu
$$
    \frac{\partial u}{\partial t} = u_t.
$$    

## 1. Klasifikasi PDE

Klasifikasi klasik PDE didasarkan pada bentuk umum PDE linier orde dua yang tampak seperti
$$
    a u_{xx} + 2b u_{xy} + c u_{yy} + d u_x + e u_y + f u = g.
$$

Perlu dicatat bahwa dalam kerangka umum klasifikasi $x$ atau $y$ bisa juga $t$ atau variabel lain. Secara formal mengambil transformasi Fourier dari 
$$
    a u_{xx} + 2b u_{xy} + c u_{yy} + d u_x + e u_y + f u = g.
$$
memungkinkan kita untuk mengubah persamaan umum ini menjadi
$$
    P(X,Y) = a X^2 + 2b X \cdot Y + c Y^2 + \cdots
$$
yang terlihat seperti polinomial orde dua umum.

Terminologi klasifikasi PDE berasal dari kenyataan bahwa persamaan di atas menggambarkan [garis kerucut](https://en.wikipedia.org/wiki/Conic_section) umum yang sering diklasifikasikan sebagai eliptik, parabolik, atau hiperbolik berdasarkan diskriminannya
$$
    b^2 - a c.
$$
Kita kemudian mengklasifikasikan PDE sesuai dengan tanda diskriminan sehingga
 - jika $b^2 - a c < 0$ PDE adalah eliptik,
 - jika $b^2 - a c = 0$ PDE adalah parabolik, dan
 - jika $b^2 - a c > 0$ PDE adalah hiperbolik.

Alasan kita peduli dengan klasifikasi adalah karena sering terdapat kelas metode numerik yang cocok untuk satu klasifikasi tetapi tidak untuk yang lain. Kita akan menyelidiki lebih lanjut mengapa dan bagaimana ini bekerja saat kita memperkenalkan setiap jenis PDE tetapi untuk saat ini mari kita periksa beberapa contoh dasar PDE dari setiap klasifikasi.

### Contoh

**Persamaan Laplace:** Dalam kasus berdimensi $n$ persamaan Laplace adalah
$$
    \Delta u = \nabla^2 u = \sum^n_{i=1} u_{x_i x_i} = 0
$$
dan ketika $n = 2$ menggunakan $x$ dan $y$ sebagai dimensi kita memiliki
$$
    u_{xx} + u_{yy} = 0.
$$
PDE jenis apa ini?



Mengidentifikasi koefisien di atas sebagai $a = 1$, $b = 0$, dan $c = 1$ maka diskriminan memberi tahu kita bahwa persamaan Laplace bersifat eliptik karena
$$
    b^2 - a c = -1.
$$

**Persamaan Panas:** Persamaan panas didefinisikan dalam satu dimensi ruang sebagai
$$
    u_t = u_{xx}.
$$

PDE jenis apa persamaan panas ini?

Persamaan panas adalah parabolik. Hal ini dapat dilihat dengan mengganti turunan $y$ di atas dengan $t$ dan dengan demikian menemukan koefisien $a=-1$, $b=0$, dan $c=0$ yang berarti diskriminan adalah nol.

Bagaimana dengan persamaan panas dalam beberapa dimensi ruang?
$$
    u_t = u_{xx} + u_{yy}.
$$
Sayangnya identifikasi ini runtuh di sini karena kita memiliki tiga variabel.

**Persamaan Gelombang:** Persamaan gelombang dalam satu dimensi ruang didefinisikan sebagai
$$
    u_{tt} - u_{xx} = 0.
$$

Dengan proses eliminasi Anda mungkin menebak bahwa persamaan ini adalah hiperbolik. Lagi mengganti turunan $y$ dengan $t$ kita menemukan $a = -1$, $b = 0$, $c = 1$ sehingga

$$
    b^2 - a c = -(-1)(1) = 1.
$$

**Persamaan Navier-Stokes**: Model aliran fluida inkompresibel
$$\begin{aligned}
    u_t + (u^2)_x + (uv)_y = - \frac{1}{\rho} p_x + \nu (u_{xx} + v_{yy})\\
    v_t + (uv)_x + (v^2)_y = - \frac{1}{\rho} p_y + \nu (u_{xx} + v_{yy})
\end{aligned}$$

Bahkan mengabaikan fakta bahwa persamaan ini tidak dapat diklasifikasikan melalui nomenklatur kita, mereka mengandung sesuatu yang tampak seperti bagian hiperbolik (sisi kiri) dan bagian eliptik (sisi kanan) atau parabolik mengingat turunan waktu. Karena hal ini persamaan Navier-Stokes sering dianggap parabolik namun pendekatan untuk menyelesaikannya sering menggunakan teknik dari masing-masing klasifikasi.

Salah satu cara yang lebih berguna untuk memikirkan klasifikasi PDE yang lebih umum adalah dengan memeriksa bagaimana informasi merambat melalui domain.  

 - **PDE Hiperbolik:** informasi bergerak pada kecepatan terbatas (kecepatan grup nyata dan terbatas).  
 - **PDE Parabolik:** menggambarkan pergerakan informasi yang tak terbatas (pikirkan apa yang terjadi dalam kasus difusi di vakum).  
 - **PDE Eliptik:** adalah kasus batas antara persamaan hiperbolik dan eliptik dalam hal propagasi informasi.

## 2. Analisis Fourier

Analisis Fourier adalah salah satu alat dasar untuk menganalisis dan menyelesaikan PDE linier. Untuk domain terbatas kita biasanya menggunakan deret Fourier untuk merepresentasikan solusi. Penggunaan deret Fourier setara dengan mempertimbangkan fungsi eigen dari operator diferensial sebagai basis solusi. Interpretasi aljabar linear PDE ini sangat kuat dan akan kita bahas berulang kali.

### Deret Fourier

Deret Fourier sebuah fungsi hadir dalam beberapa variasi tergantung pada jenis fungsi yang kita pertimbangkan. Ingat bahwa $\sin x$ adalah fungsi ganjil dan $\cos x$ adalah fungsi genap. Pada interval terbatas maka untuk fungsi ganjil kita dapat menggunakan deret $\sin$ Fourier dan untuk fungsi genap kita dapat menggunakan deret $\cos$ Fourier. Jika fungsi tidak ganjil maupun genap maka kita dapat menggunakan deret lengkap, sering ditulis sebagai
$$
    f(x) = a_0 + \sum^\infty_{n=1} a_n \cos \left( \frac{n \pi x}{L} \right) + \sum^\infty_{n=1} b_n \sin \left( \frac{n \pi x}{L} \right),
$$
di mana
$$\begin{aligned}
    a_0 &= \frac{1}{L} \int^L_0 f(x) dx, \\
    a_n &= \frac{2}{L} \int^L_0 f(x) \cos \frac{n \pi x}{L} dx, \\
    b_n &= \frac{2}{L} \int^L_0 f(x) \sin \frac{n \pi x}{L} dx.
\end{aligned}$$
Koefisien-koefisien ini adalah proyeksi fungsi $f(x)$ ke basis Fourier dari $\cos$ dan $\sin$ yang membentuk basis ortogonal.

Bagaimana ini berkaitan dengan aljabar linear dan ruang vektor berdimensi hingga?

### Transformasi Fourier

Transformasi Fourier sebuah fungsi $v(x)$ didefinisikan oleh
$$
    \widehat{v}(\xi) = \frac{1}{\sqrt{2 \pi}} \int^\infty_{-\infty} v(x) e^{-i \xi x} dx.
$$
Agar transformasi ada kita memerlukan bahwa fungsi $v \in L^2$, yaitu
$$
    ||v||_2 = \left ( \int^\infty_{-\infty} \left|v(x) \right|^2 dx \right )^{1/2} < \infty.
$$
Fungsi $\widehat{v}(\xi) \in L^2$ karena dari identitas Parseval kita memiliki
$$
    ||v(x)||_2 = ||\widehat{v}(\xi)||_2.
$$
Kita juga dapat menggunakan transformasi invers untuk mendapatkan kembali fungsi asli:
$$
    v(x) = \frac{1}{\sqrt{2 \pi}} \int^\infty_{-\infty} \widehat{v}(\xi) e^{i \xi x} d\xi.
$$

Salah satu sifat terpenting dari Transformasi Fourier yang telah kita gunakan di atas adalah bahwa transformasi Fourier mengubah turunan spasial menjadi hubungan aljabar:
$$
    \widehat{v_x} = i \xi \widehat{v}.
$$
Ini adalah hasil dari fungsi 
$$
    e^{i\xi x}
$$ 
menjadi fungsi eigen dari operator
$$
    \frac{\partial}{\partial x}
$$
Hal ini juga diterjemahkan ke dalam kasus diskrit di mana kita akan menemukan bahwa analog diskrit dari fungsi eigen tersebut akan menjadi vektor eigen dari operator beda hingga yang setara.

**Latihan:** Cobalah mengambil transformasi Fourier dari persamaan adveksi:
$$
    u_t + c u_x = 0.
$$

Karena turunan waktu tidak terpengaruh oleh transformasi Fourier kita berakhir dengan
$$
    \widehat{u}_t = -i \xi c \widehat{u}
$$
yang merupakan sistem ODE tak hingga. Sekarang cobalah menyelesaikan sistem ini untuk menemukan solusi persamaan adveksi.

Kita tahu solusi ODE tersebut berbentuk
$$
    \widehat{u}(\xi, t) = \widehat{u}(\xi, 0) e^{-i \xi c t}.
$$
Menggunakan transformasi invers kita menemukan
$$\begin{aligned}
    u(x,t) &= \frac{1}{\sqrt{2 \pi}} \int^\infty_{-\infty} \widehat{u}(\xi, 0)  e^{-i \xi c t} e^{i \xi x} d\xi, \\
    &= \frac{1}{\sqrt{2 \pi}} \int^\infty_{-\infty} \widehat{u}(\xi, 0)  e^{i \xi (x - c t)} d\xi, \\
    &= u(x - c t, 0).
\end{aligned}$$

## 3. Persamaan Eliptik

Persamaan eliptik paling sering merupakan masalah nilai batas pada suatu domain. Di atas kita menyebutkan persamaan Laplace yang merupakan kasus khusus dari persamaan eliptik klasik lainnya, masalah Poisson
$$
    \nabla^2 u = f.
$$
Jenis masalah ini sering ditemukan ketika kita ingin mencari keadaan mantap dari beberapa masalah bergantung waktu seperti persamaan panas. Ini juga dapat muncul dalam masalah bergantung waktu di mana terdapat beberapa dinamika yang jauh lebih cepat daripada skala waktu bagian lain dari masalah, salah satunya adalah kondisi inkompresibilitas yang digunakan dalam persamaan Navier-Stokes (masalah Poisson untuk tekanan).

Perilaku esensial dari masalah nilai batas eliptik adalah bahwa solusi di interior domain $\Omega$ ditentukan sepenuhnya oleh kondisi batas dan fungsi $f$. Ini penting ketika mempertimbangkan metode numerik untuk jenis masalah ini karena kita sering harus menyelesaikan sistem persamaan linear besar dan jarang yang menghubungkan setiap titik ke setiap titik lainnya.

Secara umum operator linier $L$ berbentuk
$$
    L = \sum^n_{i,j=1} A_{ij} \frac{\partial^2}{\partial x_i \partial x_j} + \sum^n_{i=1} B_i \frac{\partial}{\partial x_i} + C
$$
dapat disebut eliptik jika matriks $A$ positif atau negatif definit (yaitu $v^T A v$ memiliki tanda yang sama untuk $v \neq 0$ dan $v \in \mathbb{R}^n$).

### Menyelesaikan Persamaan Eliptik

Sebagai contoh penyelesaian PDE eliptik mari kita pertimbangkan persamaan Laplace pada domain persegi panjang
$$\begin{aligned}
    &u_{xx} + u_{yy} = 0 \quad \Omega = (0, L) \times (0, H) \\
    &u(x,0) = f_1(x) \\
    &u(x,H) = f_2(x) \\
    &u(0,y) = g_1(y) \\
    &u(L,y) = g_2(y).
\end{aligned}$$

Karena kita memiliki kondisi batas tidak homogen pada setiap sisi persegi panjang kita tidak dapat langsung menggunakan pemisahan variabel dan sebagai gantinya memecah masalah menjadi 4 subbagian sehingga
$$
    u(x,y) = u_1(x,y) + u_2(x,y) + u_3(x,y) + u_4(x,y)
$$
di mana setiap $u_i(x,y)$ mempertahankan salah satu sisi persegi panjang yang tidak homogen dan lainnya dijaga nol. Misalnya untuk $u_1(x,y)$ kita memiliki
$$\begin{aligned}
    \nabla^2 u_1 = 0 \quad & \quad \Omega = (0, L) \times (0, H) \\
    u_1(x,0) &= f_1(x) \\
    u_1(x,H) &= 0 \\
    u_1(0,y) &= 0 \\
    u_1(L,y) &= 0.
\end{aligned}$$

Menggunakan pemisahan variabel kita mengasumsikan solusi berbentuk $u_4(x,y) = h(x) \phi(y)$ yang disubstitusikan ke PDE menghasilkan
$$\begin{aligned}
    &\phi \frac{\text{d}^2 h}{\text{d} x^2} + h \frac{\text{d}^2 \phi}{\text{d} y^2} = 0 \\
    &\frac{1}{h} \frac{\text{d}^2 h}{\text{d} x^2} = - \frac{1}{\phi} \frac{\text{d}^2 \phi}{\text{d} y^2} = -\lambda \Rightarrow \\
    &h''(x) + \lambda h(x) = 0 \quad\quad \phi''(y) - \lambda \phi(y) = 0.
\end{aligned}$$

Kondisi batas untuk kedua fungsi adalah
$$
    h(0) = h(L) = 0 \quad \phi(0) = f_1(x) \quad \phi(H) = 0.
$$
Pertama menyelesaikan BVP homogen yang lebih mudah untuk $h(x)$ kita menemukan
$$
    h(x) = A \cos (\sqrt{\lambda} x) + B \sin(\sqrt{\lambda} x)
$$
yang menggunakan kondisi batas membawa kita ke
$$\begin{aligned}
    &h(0) = 0 = A \\
    &h(L) = 0 = B \sin(\sqrt{\lambda} L).
\end{aligned}$$

Yang kedua kita memiliki dua pilihan untuk memenuhi kondisi batas, kita dapat mengasumsikan $B=0$ yang mengarah ke solusi trivial atau
$$
    \sqrt{\lambda} L = n \pi.
$$
Menggunakan kondisi terakhir menghasilkan nilai eigen
$$
    \lambda_n = \left(\frac{n \pi}{L}\right)^2, \quad n = 1,2,\ldots
$$
sehingga kita memiliki solusi
$$
    h_n(x) = B_n \sin\left(\frac{n \pi x}{L}\right).
$$

Untuk persamaan $\phi$ kita memiliki solusi berbentuk
$$
    \phi(y) = A e^{\sqrt{\lambda} y} + B e^{-\sqrt{\lambda} y}
$$
atau
$$
    \phi(y) = A \cosh(\sqrt{\lambda} y) + B \sinh(\sqrt{\lambda} y).
$$

Untuk memenuhi kondisi batas homogen kita perlu memodifikasi solusi ini sedikit sehingga digeser
$$
    \phi(y) = A \cosh \left(\sqrt{\lambda} (y - H)\right) + B \sinh \left(\sqrt{\lambda} (y - H)\right)
$$
yang memungkinkan kita memenuhi kondisi batas homogen dengan
$$\begin{aligned}
    \phi(H) = 0 &= A \cosh \left(\sqrt{\lambda} (H - H)\right) + B \sinh \left(\sqrt{\lambda} (H - H)\right) \\
    &=A \cosh (0) + B \sinh (0) \\
    &=A
\end{aligned}$$
dan kita kemudian tersisa dengan
$$
    \phi(y) = B \sinh \left(\frac{n \pi}{L} (y - H)\right)
$$

Menggabungkan kedua solusi kita kemudian memiliki
$$
    u_{1,n}(x,y) = B_n \sin\left(\frac{n \pi x}{L}\right) \sinh \left(\frac{n \pi}{L} (y - H)\right)
$$
dan dengan superposisi kita punya
$$
    u_1(x,y) = \sum^\infty_{i=1} B_n \sin\left(\frac{n \pi x}{L}\right) \sinh \left(\frac{n \pi}{L} (y - H)\right)
$$

Sekarang, untuk memenuhi kondisi batas terakhir dan menemukan koefisien $B_n$ kita memiliki
$$
    u_1(x,0) = f_1(x) = \sum^\infty_{i=1} -B_n \sin\left(\frac{n \pi x}{L}\right) \sinh \left(\frac{n \pi H} {L}\right)
$$
dan menggunakan ortogonalitas fungsi $\sin$ kita memiliki
$$\begin{aligned}
    \int^L_0 f_1(x) \sin\left(\frac{m \pi x}{L}\right) dx &= \sum^\infty_{i=1} -B_n \sinh \left(\frac{n \pi H}{L} \right ) \int^L_0 \sin\left(\frac{n \pi x}{L}\right) \sin\left(\frac{m \pi x}{L}\right) dx \\
    \int^L_0 f_1(x) \sin\left(\frac{m \pi x}{L}\right) dx &= -B_m \frac{L}{2} \sinh \left(\frac{n \pi H}{L} \right ) \\
    B_m &= \frac{-2}{L \sinh \left(\frac{m \pi H}{L} \right )} \int^L_0 f_1(x) \sin\left(\frac{m \pi x}{L}\right) dx.
\end{aligned}$$
Kita kemudian telah menemukan solusi untuk komponen $u_1$ dari masalah umum di atas.

## 4. Persamaan Parabolik

Persamaan parabolik mirip dengan persamaan eliptik namun sering juga mencakup ketergantungan waktu (ini tidak selalu terjadi tetapi akan demikian untuk pertimbangan kita). Kita misalnya dapat mengambil operator linier eliptik apa pun seperti yang didefinisikan di atas dan membuat persamaan parabolik dengan menulis
$$
    u_t - Lu = f.
$$

Contoh klasik persamaan parabolik adalah persamaan panas (atau difusi) yang didefinisikan sebagai
$$
    u_t = \nabla^2 u.
$$
PDE ini memerlukan kondisi awal $u(x, 0)$ serta kondisi batas pada $\Omega$.

### Menyelesaikan Persamaan Parabolik

Mirip dengan masalah eliptik kita kembali akan menggunakan deret Fourier untuk menyelesaikan persamaan parabolik. Mari kita kembali mempertimbangkan masalah model kita, persamaan panas, dalam satu dimensi ruang:
$$\begin{aligned}
    &u_t = \kappa u_{xx}, \quad \Omega = (0, T) \times (0, L), \\
    &u(x, 0) = f(x), \\
    &u(0, t) = 0, \quad \text{dan} \\
    &u(L, t) = 0.
\end{aligned}$$

Menggunakan pemisahan variabel kita asumsikan $u(x,t) = G(t) \phi(x)$ dan kemudian dapat menulis dua ODE
$$\begin{aligned}
    &\frac{\text{d} G}{\text{d} t} = -\kappa \lambda G, \\
    &\phi''(x) + \lambda \phi(x) = 0.
\end{aligned}$$
ODE tergantung waktu memiliki solusi bentuk
$$
    G(t) = C e^{-\lambda \kappa t}.
$$
BVP untuk $\phi$ memiliki solusi bentuk
$$
    \phi(x) = A \cos \left( \sqrt{\lambda} x \right) + B \sin \left( \sqrt{\lambda} x \right).
$$

Menggunakan kondisi batas kita menemukan bahwa $A = 0$ dan untuk solusi non-trivial
$$
    \lambda = \left( \frac{n \pi}{L} \right )^2,
$$
yang menghasilkan solusi bentuk
$$
    u_n(x,t) = B_n e^{-\left(\frac{n \pi}{L}\right)^2 \kappa t} \sin \left( \frac{n \pi x}{L} \right),
$$
dan dengan superposisi kita memiliki
$$
    u(x,t) = \sum^\infty_{i=1} B_n e^{-\left(\frac{n \pi}{L}\right)^2 \kappa t} \sin \left( \frac{n \pi x}{L} \right).
$$

Mengevaluasi kondisi awal kita memiliki
$$
    u(x,0) = f(x) = \sum^\infty_{i=1} B_n \sin \left( \frac{n \pi x}{L} \right),
$$
yang memberikan kita nilai $B_n$ dengan kembali menggunakan ortogonalitas fungsi $\sin$
$$\begin{aligned}
    \int^L_0 f(x) \sin \left( \frac{m \pi x}{L} \right) dx &= \sum^\infty_{i=1} B_n \int^L_0\sin \left( \frac{n \pi x}{L} \right) \sin \left( \frac{m \pi x}{L} \right) dx, \\
    \int^L_0 f(x) \sin \left( \frac{m \pi x}{L} \right) dx &= B_m \frac{L}{2}, \\
    B_m &= \frac{2}{L} \int^L_0 f(x) \sin \left( \frac{m \pi x}{L} \right) dx,
\end{aligned}$$
sehingga menyelesaikan PDE.

## 5. Persamaan Hiperbolik

Persamaan hiperbolik sering muncul sebagai hukum konservasi atau keseimbangan. Misalnya konservasi momentum (hukum kedua Newton) adalah contoh hukum keseimbangan. Demikian juga konservasi energi dapat ditulis sebagai persamaan hiperbolik. Persamaan hiperbolik klasik tentu saja adalah persamaan gelombang
$$
    u_{tt} - c^2 u_{xx} = 0.
$$
Namun sering ketika mempertimbangkan metode numerik untuk persamaan hiperbolik kita menguraikan PDE orde dua menjadi sistem PDE orde satu seperti
$$
    u_t + A u_x = 0,
$$
di mana kini $A$ adalah matriks. Dalam kasus ini PDE bersifat hiperbolik jika nilai eigen matriks $A$, $\lambda$, bersifat real dan matriks dapat diagonal.

### Menyelesaikan Persamaan Hiperbolik

Pendekatan paling umum untuk menyelesaikan PDE hiperbolik adalah menemukan karakteristik dari masalah yang diberikan dan menggambar titik-titik di bidang ($x$,$t$) di mana sesuatu yang menarik mungkin terjadi. Jika persamaan non-linear hal ini dapat menyebabkan kejutan (diskontinuitas) meskipun kondisi awalnya halus. Ini akan menjadi masalah bagi kita secara numerik seperti yang akan kita lihat nanti dalam kursus.

Pendekatan dasar metode karakteristik menggunakan asumsi bahwa solusi kita berbentuk
$$
    u(x,t) = u(x(t), t),
$$
atau dengan kata lain solusi kita dapat digambarkan oleh satu variabel karakteristik, dalam hal ini $t$. Misalkan kita mencoba mencari solusi untuk
$$
    u_t + c(x,t,u) u_x = 0.
$$

Jika kita mengambil turunan penuh (atau total) dari solusi baru kita mendapatkan
$$
    \frac{\text{d} u}{\text{d} t} = \frac{\partial u}{\partial t} + \frac{\text{d} x}{\text{d} t} \frac{\partial u}{\partial x},
$$
yang, setelah mentransformasikan persamaan asli untuk menggunakan $\xi$, membawa kita pada kesimpulan bahwa PDE asli kita dapat ditulis ulang menjadi dua ODE
$$\begin{aligned}
    &\frac{\text{d} u}{\text{d} t} = 0, \quad \text{dan}\\
    &\frac{\text{d} x}{\text{d} t} = c(x,t,u).
\end{aligned}$$
Sekarang kita tahu bahwa $u(x(t),t)$ konstan sepanjang kurva $x(t)$ dan sebenarnya
$$
    u(x,t) = u(x(t), 0).
$$

Solusi PDE kemudian berubah menjadi menyelesaikan kurva karakteristik yang diberikan oleh $x(t)$.

Sebagai contoh sederhana mari selesaikan persamaan adveksi linier pada domain tak hingga
$$\begin{aligned}
    &u_t + c u_x = 0, \quad \Omega = (-\infty, \infty) \times (0, T) \\
    &u(x,0) = f(x), \quad \text{dan} \\
    &u_x(\pm \infty, t) \rightarrow 0.
\end{aligned}$$
Persamaan karakteristik adalah
$$
    \frac{\text{d} x}{\text{d} t} = c,
$$
yang solusinya adalah $x(t) = c t + x_0$. Kita dapat menggambar karakteristik tersebut sebagai

![Kecepatan Gelombang Konstan](./images/characteristics.png)

di mana persamaan garisnya adalah $x(t) = c t + x_0$.

Untuk persamaan nonlinier, yaitu yang kecepatan mungkin bergantung pada solusi $c(u)$, kondisi awal akan menentukan kemiringan karakteristik yang mengarah ke dua kasus. Mari kita pertimbangkan persamaan Burger
$$
    u_t + u u_x = 0,
$$
dengan pengaturan yang sama seperti PDE linear di atas. Cobalah untuk melihat apakah Anda dapat menentukan kurva karakteristik untuk kondisi awal
$$
    u(x,0) = \left \{ \begin{aligned}
        4, \;\; & x<0, \\
        3, \;\; & x>0.
    \end{aligned} \right. 
$$
Apa karakteristik untuk masalah ini? Apa yang menjadi masalah dalam kasus ini?

Kurva karakteristik akan menjadi
$$
    x(t) = \left \{ \begin{aligned}
        4 t + x_0, \;\; & x<0, \\
        3 t + x_0, \;\; & x>0,
    \end{aligned} \right. 
$$
yang mengarah pada tabrakan kurva karakteristik dan karena itu sebuah shock.

![Shock](./images/characteristics_shock.png)

Kita dapat menemukan persamaan untuk lokasi shock dengan menggunakan kondisi Rankine-Hugoniot
$$
    \frac{\text{d} x_s}{\text{d} t} = \frac{f(x_+, t) - f(x_-, t)}{u_+ - u_-},
$$
di mana $x_s(t)$ adalah lokasi shock, $f$ adalah fungsi fluks yang didefinisikan dari $u_t + f(u)_x = 0$, dan notasi $u_\pm$ menunjukkan nilai di kiri dan kanan shock tergantung pada tanda.

Bagaimana jika kita memiliki kondisi awal
$$
    u(x,0) = \left \{ \begin{aligned}
        3, \;\; & x<0, \\
        4, \;\; & x>0,
    \end{aligned} \right.
$$
sebagai gantinya? Jenis solusi apa yang sekarang kita harapkan?

Kurva karakteristik sekarang
$$
    x(t) = \left \{ \begin{aligned}
        3 t + x_0, \;\; & x<0, \\
        4 t + x_0, \;\; & x>0,
    \end{aligned} \right. 
$$
meninggalkan bagian bidang $(x,t)$ tanpa karakteristik. Sebaliknya kita mengasumsikan solusi yang bervariasi halus, disebut rarefaksi, dan mengisi solusi yang hilang.

![Rarefaksi](./images/characteristics_rarefaction.png)

Dalam kasus ini kita ingin menemukan solusi yang bervariasi halus di dalam wedge ini yang kita temukan dengan mengasumsikan bahwa solusi bersifat swakelar. Ini menyiratkan bahwa $u$ berbentuk
$$
    u(x, t) = u(\xi),
$$
di mana untuk PDE hiperbolik orde satu variabel swakelar adalah
$$
    \xi = x / t.
$$
Cobalah memasukkan ini ke dalam persamaan asli maka akan memberi kita solusi dalam bentuk $u(\xi)$ untuk interior wedge.

$$\begin{aligned}
    &u_t + u u_x = 0 \\
    &u' \frac{\partial \xi}{\partial t} + u u' \frac{\partial \xi}{\partial x} = 0 \\
    &u' \left ( -\frac{x}{t^2} + u \frac{1}{t} \right ) = 0 \\
    &u' \left ( -\xi + u\right ) = 0
\end{aligned}$$
Persamaan terakhir ini sekarang dapat diselesaikan.
