## 1. Deret Fourier lengkap (Full Fourier series pada interval [0, L])

$$ f(x) = a_0 + \sum_{n=1}^\infty a_n \cos\left( \frac{n\pi x}{L} \right) + \sum_{n=1}^\infty b_n \sin\left( \frac{n\pi x}{L} \right) $$

**Penjelasan:**  
Ini adalah representasi umum suatu fungsi f(x) di interval [0, L] menggunakan deret trigonometri.  
- Suku a₀ adalah komponen konstan (nilai rata-rata)  
- Deret kosinus menangkap bagian genap (even)  
- Deret sinus menangkap bagian ganjil (odd)  
Deret ini merupakan dasar untuk menyelesaikan PDE linear dengan kondisi batas periodik atau pada domain terbatas.

## 2. Koefisien konstanta a₀

$$ a_0 = \frac{1}{L} \int_0^L f(x) \, dx $$

**Penjelasan:**  
Merupakan nilai rata-rata fungsi f(x) di seluruh interval [0, L].  
Faktor 1/L muncul karena ini adalah integral rata-rata (bukan integral biasa).

## 3. Koefisien kosinus aₙ

$$ a_n = \frac{2}{L} \int_0^L f(x) \cos\left( \frac{n\pi x}{L} \right) \, dx \quad (n = 1,2,3,\dots) $$

**Penjelasan:**  
Ini adalah koefisien proyeksi fungsi f(x) ke basis fungsi cos(nπx/L).  
Faktor 2/L berasal dari hasil integral ortogonalitas: ∫ cos²(nπx/L) dx dari 0 ke L = L/2.

## 4. Koefisien sinus bₙ

$$ b_n = \frac{2}{L} \int_0^L f(x) \sin\left( \frac{n\pi x}{L} \right) \, dx \quad (n = 1,2,3,\dots) $$

**Penjelasan:**  
Koefisien proyeksi ke basis fungsi sin(nπx/L).  
Jika f(x) adalah fungsi ganjil, maka semua aₙ = 0 (hanya deret sinus yang tersisa).

## 5. Transformasi Fourier maju (Forward Fourier Transform)

$$ \widehat{v}(\xi) = \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty} v(x) e^{-i \xi x} \, dx $$

**Penjelasan:**  
Mengubah fungsi v(x) dari domain spasial (x) ke domain frekuensi (ξ).  
Konvensi ini menggunakan faktor simetris 1/√(2π) sehingga transformasi maju dan balik hampir identik bentuknya.  
Digunakan untuk fungsi di seluruh garis real ℝ.

## 6. Norma L² (syarat keberadaan transformasi Fourier)

$$ \|v\|_2 = \left( \int_{-\infty}^{\infty} |v(x)|^2 \, dx \right)^{1/2} < \infty $$

**Penjelasan:**  
Ini adalah norma L² — ukuran “energi” fungsi.  
Transformasi Fourier didefinisikan dengan baik dan bersifat isometri hanya pada ruang L²(ℝ).

## 7. Identitas Parseval / Teorema Plancherel

$$ \|v(x)\|_2 = \|\widehat{v}(\xi)\|_2 $$

**Penjelasan:**  
Energi di domain waktu sama dengan energi di domain frekuensi.  
Ini menunjukkan bahwa transformasi Fourier adalah transformasi uniter (mempertahankan norma L²).

## 8. Transformasi Fourier balik (Inverse Fourier Transform)

$$ v(x) = \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty} \widehat{v}(\xi) e^{i \xi x} \, d\xi $$

**Penjelasan:**  
Rumus untuk mengembalikan fungsi asli dari spektrum frekuensinya.  
Tanda eksponensial dibalik dibandingkan transformasi maju.

## 9. Sifat turunan spasial dalam transformasi Fourier

$$ \widehat{v_x} = i \xi \widehat{v} $$

**Penjelasan:**  
Ini adalah sifat paling penting dalam analisis PDE:  
turunan terhadap x di domain spasial menjadi operasi perkalian dengan iξ di domain frekuensi.  
Sifat inilah yang membuat transformasi Fourier sangat kuat untuk menyelesaikan persamaan diferensial linear dengan koefisien konstan.

## 10. Persamaan adveksi (persamaan transport)

$$ u_t + c u_x = 0 $$

**Penjelasan:**  
Persamaan PDE orde satu yang paling sederhana.  
Menyatakan bahwa profil fungsi u bergerak dengan kecepatan konstan c tanpa mengalami distorsi bentuk.

## 11. Bentuk transformasi Fourier dari persamaan adveksi

$$ \widehat{u}_t = -i \xi c \, \widehat{u} $$

**Penjelasan:**  
Setelah transformasi Fourier terhadap variabel spasial x, turunan uₓ menjadi iξ û.  
Persamaan PDE berubah menjadi ODE biasa untuk setiap nilai frekuensi ξ secara independen.

## 12. Solusi ODE di domain frekuensi

$$ \widehat{u}(\xi, t) = \widehat{u}(\xi, 0) \, e^{-i \xi c t} $$

**Penjelasan:**  
Solusi eksponensial dari persamaan diferensial biasa ûₜ = −i c ξ û.  
Faktor e^{-i ξ c t} menyebabkan pergeseran fase yang sebanding dengan frekuensi dan waktu.

## 13. Solusi persamaan adveksi melalui transformasi invers Fourier

$$ \begin{aligned}
u(x,t) &= \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty} \widehat{u}(\xi, 0) \, e^{-i \xi c t} \, e^{i \xi x} \, d\xi \\
&= \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty} \widehat{u}(\xi, 0) \, e^{i \xi (x - c t)} \, d\xi \\
&= u(x - c t, 0)
\end{aligned} $$

**Penjelasan langkah demi langkah:**  
1. Substitusi solusi ODE ke rumus transformasi invers  
2. Menggabungkan eksponen → e^{iξ(x − ct)}  
3. Mengidentifikasi bahwa integral tersebut sama dengan nilai fungsi awal yang digeser sebesar ct  
**Kesimpulan:** solusi persamaan adveksi adalah pergeseran murni profil awal dengan kecepatan c.
